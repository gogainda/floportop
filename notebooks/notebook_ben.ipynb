```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
import pandas as pd
import ast
```

```python
def load_csv(file_path):
    return kagglehub.load_dataset(
        KaggleDatasetAdapter.PANDAS,
        "rounakbanik/the-movies-dataset",
        file_path,
    )

movies = load_csv("movies_metadata.csv")
credits = load_csv("credits.csv")
keywords = load_csv("keywords.csv")
links = load_csv("links.csv")
links_small = load_csv("links_small.csv")
```

```python
movies = movies[movies["id"].str.isnumeric()]
movies["id"] = movies["id"].astype(int)

credits["id"] = credits["id"].astype(int)
keywords["id"] = keywords["id"].astype(int)
```

```python
links["imdbId"] = links["imdbId"].apply(lambda x: f"tt{x:07d}")
links_small["imdbId"] = links_small["imdbId"].apply(lambda x: f"tt{x:07d}")
```

```python
df = (
    movies
    .merge(credits, on="id", how="left")
    .merge(keywords, on="id", how="left")
    .merge(links, left_on="id", right_on="tmdbId", how="left")
)
df.head()
```

```python
def parse_json(col):
    return col.apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])
```

```python
df["genres"] = parse_json(df["genres"])
df["keywords"] = parse_json(df["keywords"])
df["cast"] = parse_json(df["cast"])
df["crew"] = parse_json(df["crew"])
```

## NEW STUFF INSERTED BY BEN FOR FEATURE EXTRACTION FROM WIDE_DF OVERVIEW COLUMN STARTS BELOW THIS CELL, EVERYTHING ABOVE FROM IGORS NOTEBOOK ##

```python
# Start by making sure the overview column is properly cleaned

def clean_overview(text: str) -> str:

    """ Basic cleaning step for plot overviews. Keeps meaning intact, only normalizes formatting."""

    import re
    if text is None:
        return ""
    text = str(text).lower()
    text = re.sub(r"\s+", " ", text).strip()
    return text


wide_df["overview_clean"] = wide_df["overview"].apply(clean_overview)
```

```python
# Check the current output

wide_df["overview_clean"].head(5)
```

```python
# TF-IDF is used to transform plot descriptions into weighted text features.

from sklearn.feature_extraction.text import TfidfVectorizer

def build_tfidf_overviews(text_series, max_features=8000, min_df=10):

    """ Builds TF-IDF features from plot overviews. Used to extract important plot terms per movie."""

    vectorizer = TfidfVectorizer(
        stop_words="english",
        ngram_range=(1, 2),
        min_df=min_df,
        max_features=max_features
    )
    X_tfidf = vectorizer.fit_transform(text_series)
    return X_tfidf, vectorizer


X_overview, tfidf_vectorizer = build_tfidf_overviews(wide_df["overview_clean"])
tfidf_feature_names = tfidf_vectorizer.get_feature_names_out()

# Cell runtime on my machine: 4.7s
```

```python
# The top TF-IDF terms are used to summarize each movies plot

import numpy as np

def extract_top_plot_terms(X_tfidf, feature_names, row_idx, top_k=8):

    """ Returns the most relevant TF-IDF terms for one movie plot. Used later for simple explainability."""

    row = X_tfidf[row_idx].toarray().ravel()
    if row.sum() == 0:
        return []
    top_idx = row.argsort()[-top_k:][::-1]
    return [feature_names[i] for i in top_idx if row[i] > 0]


wide_df["plot_main_points"] = [
    extract_top_plot_terms(X_overview, tfidf_feature_names, i)
    for i in range(len(wide_df))
]

# Cell runtime on my machine: 31.4s
```

```python
# LDA is used to group movies into recurring plot topics based on their descriptions

from sklearn.decomposition import LatentDirichletAllocation

RANDOM_STATE = 1

def fit_lda_topics(X_tfidf, n_topics=14, random_state=RANDOM_STATE):
    """ Fit LDA on TF-IDF plot features and return topic probabilities per movie. """
    lda = LatentDirichletAllocation(
        n_components=n_topics,
        random_state=random_state
    )
    topic_dist = lda.fit_transform(X_tfidf)
    return lda, topic_dist


lda_model, topic_dist = fit_lda_topics(X_overview, n_topics=14)

# Each movie is assigned its most likely topic along with the corresponding confidence score. Not sure if we will really need this but nice to have for now
wide_df["dominant_topic"] = topic_dist.argmax(axis=1)
wide_df["topic_confidence"] = topic_dist.max(axis=1)

# Cell runtime on my machine: 36.3s
```
